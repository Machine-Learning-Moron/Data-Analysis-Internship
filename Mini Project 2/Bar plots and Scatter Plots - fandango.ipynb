```python
import pandas as pd
fandango = pd.read_csv('fandango_scores.csv')
```

```python
fandango.head()
```

```python
norm_reviews = fandango[["FILM", 'RT_user_norm', "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]].copy()
```

```python
norm_reviews.head()
```

```python
#barplots
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import arange
num_columns = ['RT_user_norm',"Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
bar_heights = norm_reviews[num_columns].iloc[0].values
bar_positions = arange(5) + 0.75
ax.barh(bar_positions, bar_heights, 0.5)
ax.bar(bar_positions, bar_heights, 1.5)
tick_positions = range(1,6)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron")
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_columns, rotation= 90)
plt.show()
```

```python
fig, ax = plt.subplots()
x = norm_reviews["Fandango_Ratingvalue"].values
y = norm_reviews["RT_user_norm"].values
ax.scatter(x,y)
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")
plt.show()
```
